```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import qgrid
import seaborn as sns
```

```python
qgrid.nbinstall(overwrite=True)
qgrid.set_defaults(remote_js=True, show_toolbar=True)

sns.set()

plt.rcParams['figure.figsize'] = (12, 8)
%config InlineBackend.figure_format='retina'
```

```python
df = pd.read_csv('../data/train.csv')
```

```python
def survival_rate_by_field(dataframe, field, value):
    x = dataframe.Survived[dataframe[field] == value]
    return sum(x)/float(len(x))
```

### Impact of Gender on Survivability

```python
# how did women fare?
female_df = df[df.Sex == 'female']
ages = sorted(set(female_df.Age[~female_df.Age.isnull()]))
survival_rate_by_age = [survival_rate_by_field(female_df, 'Age', age) for age in ages]

plt.scatter(ages, survival_rate_by_age)
plt.title('Women Survival Rate by Age')
plt.xlabel('Age')
plt.ylabel('Survival Rate')

z = np.polyfit(ages, survival_rate_by_age, 1)
p = np.poly1d(z)
plt.plot(ages,p(ages),"r--")
plt.show()

# now for the men
male_df = df[df.Sex == 'male']
ages = sorted(set(male_df.Age[~male_df.Age.isnull()]))
survival_rate_by_age = [survival_rate_by_field(male_df, 'Age', age) for age in ages]

plt.scatter(ages, survival_rate_by_age)
plt.title('Men Survival Rate by Age')
plt.xlabel('Age')
plt.ylabel('Survival Rate')

z = np.polyfit(ages, survival_rate_by_age, 1)
p = np.poly1d(z)
plt.plot(ages,p(ages),"r--")
plt.show()
```
